### Bitcoin Data

```python
dateparse = lambda dates: pd.datetime.strptime(dates, '%d-%m-%Y')
btc=pd.read_csv('E:\CRYPTO/Bitcoin.csv',parse_dates=['Date'],index_col='Date', date_parser=dateparse)
btc = btc.iloc[::-1]
btc.tail(5)
```

```python
maxValue=btc[btc['Close']==max(btc.Close)]
print("Highiest value of bitcoin")
maxValue
```

```python
btc.describe()
```

```python
btc.info()
```

```python
plt.figure(figsize = (20,10))
btc.groupby('Date')['Volume'].median().plot(color = 'g')
plt.xlabel('Date')
plt.ylabel('Volume')
```

```python
df1 = btc.tail(30)
```
